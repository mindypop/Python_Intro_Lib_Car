```python
#von Githun laden mit, und umbenennen um kürzer zu halten (pandas wird immer mit pd abgekürzt, steht so auch auf der pandas Seite)
import urllib.request
import pandas as pd
```

```python
#Gucken uns hier open access Dateien der Bibliotheken an (apc)
apc_url = "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2015.csv"
apc_file = "openapc.csv"
```

```python
#Lädt Datei runter. Speichert in apc_file ab.
urllib.request.urlretrieve(apc_url,apc_file)
```

```python
#Variabe zuweisen. Option mit dem Lesen pd.read muss man sich aus der Doku holen. Umgeschriebe und in Variable geschrieben.
apcs = pd.read_csv(apc_file)
```

```python
#Hier sieht man den Typ und dass es ein data frame ist.
type(apcs)
```

```python
#Wenn keine Daten: NaN. Hat Index, 0,1,2, etc. Nur die ersten 30 und letzten 30. Wird nicht alles angezeigt, ist aber da. Unten steht wie viel rows und spalten
apcs
```

```python
#Nur Spalte mit Euros ausgeben lassen.
apcs["euro"]
```

```python
#Filtern. Größer als 2000 oder kleiner
apcs["euro"] >= 2000
```

```python
#Jetzt nur die wahren anzeigen lassen (also über 2000). mit dem ersten apcs wird der 
#gesammte Frame ausgesucht, dann nur die Spalte.
apcs[apcs["euro"] >= 2000]
```

```python
#Spaltennamen ausgeben, welche gibt es:
apcs.columns
```

```python
#weiteres Attribut (wie column festgeschriebene Eigenschaften, die ausgegeben werden:
apcs.shape
```

```python
#Ist 2000 teuer oder nicht. histogramm machen (matplotlib, um Graphen im notebook hier anzeigen lassen. 
#wir nehmen Spalte euro und machen darauf diese history:
%matplotlib inline
apcs["euro"].hist()
```

```python
#Argumente: bin und Wert geben. Die Argumente findet man in der Doku. Jetzt weniger bins. wird mehr zusammengezogen. 
#Mit 10 wie oben und mit 20 feiner, etc.
apcs["euro"].hist(bins=40, color="green")
```
